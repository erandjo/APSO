## Import statements

```python
import os
import pandas as pd
from collections import Counter
```

## Loading dataset

```python
def load_dataset(): # add file path 
    path = os.getcwd() + "\\..\\..\\json-files\\Intermediate Files\\from_android_emulator.json"
    print("PATH: {}".format(path))
    return pd.read_json(path)

dataset = load_dataset()
dataset[0:3]
```

A fast inspection of the provided descriptions and categories show that these are unreliable for any type of clustering.

```python
def _inspect_column(name):
    c = Counter(dataset[name])
    return c.items()

pmCategories = _inspect_column("category")
pmDescriptions = _inspect_column("description")

print("Categories from package manager {} ".format(list(pmCategories)))
print("Descriptions from package manager {} ".format(list(pmDescriptions)))
```

We see that 49 out of 79 applications have no specified category, and none of the 79 applications have a description.

## Removing Samples

Removing my Applicaiton which was only used to do the collection; we also remove an app called QR Code because we were not able to figure out what app this was, and could thus not collect author nor category

```python
def _find_samples_to_drop():
    print(dataset[dataset["label"] == "My Application"] == True)
    print(dataset[dataset["label"] == "QR Code"] == True)
_find_samples_to_drop()
```

```python
def _drop_samples():
    df = dataset.drop(index=64)
    df = df.drop(index=22)
    df = df.reset_index()
    return df
datasetA = _drop_samples()
```

## Writing dataset to file
